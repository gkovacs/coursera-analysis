```python
import MySQLdb
import json

from pandas import DataFrame
from pandas.io.sql import read_frame

datastage_login = json.load(open('datastage_login.json'))

database = MySQLdb.connect(host='datastage.stanford.edu', user=datastage_login['user'], passwd=datastage_login['passwd'])
cursor = database.cursor()
cursor.execute('show databases')

course_start_times = {}

for (dbname,) in cursor:
  if 'anonymized_general' not in dbname:
    continue
  cursor.execute('use `' + dbname + '`')
  cursor.execute('show tables')
  tables = frozenset([tablename for (tablename,) in cursor])
  if 'lecture_metadata' in tables:
    data = read_frame('select min(open_time) from lecture_metadata', database)
    start_time = int([x for i,x in data[['min(open_time)']].itertuples()][0])
    course_start_times[dbname] = start_time

print course_start_times
```

```python
%matplotlib inline
import matplotlib.pyplot
import numpy

def barPlotWithStderr(means, stderrs=None, x_tick_labels=None, xlabel='', ylabel='', title=''):
    if stderrs:
        assert len(means) == len(stderrs)
    if x_tick_labels:
        assert len(means) == len(x_tick_labels)
        x_tick_labels = [str(x) for x in x_tick_labels]
    else:
        x_tick_labels = [str(x) for x in range(len(means))]
    N = len(means)
    ind = numpy.arange(N) # the x locations for the groups
    width = 0.35 # the width of the bars
    fig, ax = matplotlib.pyplot.subplots()
    if stderrs:
        rects1 = ax.bar(ind, means, width, color='r', yerr=stderrs)
    else:
        rects1 = ax.bar(ind, means, width, color='r')
    
    if ylabel and not title:
        title = ylabel
    ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.set_xlabel(xlabel)
    ax.set_xticks(ind+width)
    ax.set_xticklabels( x_tick_labels )
    
    def autolabel(rects):
        # attach some text labels
        for rect in rects:
            height = rect.get_height()
            #ax.text('foobar', 1.05*height, '%d'%int(height), ha='center', va='bottom')
            ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, '%d'%int(height), ha='center', va='bottom')
    
    autolabel(rects1)

def plotBarDictionary(datadict, xlabel='', ylabel='', title='', x_tick_label_map=None):
    keys = datadict.keys()
    keys.sort()
    means = [datadict[k] for k in keys]
    if not x_tick_label_map:
      x_tick_labels = keys
    else:
      x_tick_labels = [x_tick_label_map[key] for key in keys]
    barPlotWithStderr(means, x_tick_labels=x_tick_labels, xlabel=xlabel, ylabel=ylabel, title=title)
```

```python
import datetime

month_to_num_courses = {}

for course,start_time in course_start_times.iteritems():
  date = datetime.datetime.fromtimestamp(start_time)
  month = date.month
  if month not in month_to_num_courses:
    month_to_num_courses[month] = 0
  month_to_num_courses[month] += 1


x_tick_label_map = {}
for i,month in enumerate(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']):
  x_tick_label_map[i+1] = month

plotBarDictionary(month_to_num_courses, x_tick_label_map=x_tick_label_map, xlabel='Month', ylabel='Num courses started in this month', title='Months courses are started in')
```
